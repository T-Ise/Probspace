```python
# 含まれている文字列を取得する関数
def get_str(s):

    # スラッシュが存在する場合、その前を切り出し
    return np.nan if s.find('/') == -1 else s[:s.find('/')]
```

```python
# 抽出文字列を追加
train_df['str_from_id'] = train_df['id'].apply(get_str)

train_df[['id', 'str_from_id']]
```

```python
# 抜き出した文字列の度数分布
ext_str_dist = train_df['str_from_id'].value_counts(dropna=False).to_frame(name='出現度数')

# パーセンテージを記入
ext_str_dist['割合'] = ext_str_dist['出現度数'] / ext_str_dist['出現度数'].sum()

ext_str_dist
```
