## データ読み込み

```python
# 訓練データを読み込み
train_df = pd.read_pickle(data_path / 'train_df.pkl')

train_df
```

## データ精査
> None有無確認

```python
list(train_df['submitter'])
```

- Noneは存在するぽい

```python
# Noneをnanに変換しつつ、全ての度数分布を確かめる
submitter_dist = train_df['submitter'].apply(lambda x: np.nan if x is None else x).value_counts(dropna=False).to_frame(name='出現度数')

# 表示しきれないので外部に出力
submitter_dist.to_csv(result_path / ('%s_submitter_度数分布.csv' % prefix), encoding='utf-8-sig')

submitter_dist
```

- Collaborationでたくさん論文出している人、個人でたくさん論文出している人色々いる

```python
# 著者ごとに各被引用数の平均を出す
cites_mean_sub = train_df.groupby('submitter')['cites'].mean().sort_values(ascending=False).to_frame(name='有効cites平均')

cites_mean_sub
```

```python
# 著者ごとに各被引用数の平均を出す
cites_count_sub = train_df.groupby('submitter')['cites'].count().sort_values(ascending=False).to_frame(name='有効cites件数')

cites_count_sub
```

```python
# doi_citesが文字列として保存されているのでintに変換
train_df['doi_cites'] = train_df['doi_cites'].apply(lambda x: int(x))
```

```python
# 著者ごとに低精度各被引用数の平均を出す
doi_cites_mean_sub = train_df.groupby('submitter')['doi_cites'].mean().sort_values(ascending=False).to_frame(name='doi_cites平均')

doi_cites_mean_sub
```

```python
# 著者ごとに低精度各被引用数の平均を出す
doi_cites_count_sub = train_df.groupby('submitter')['doi_cites'].count().sort_values(ascending=False).to_frame(name='doi_cites件数')

doi_cites_count_sub
```
